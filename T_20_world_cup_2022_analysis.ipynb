# Correlation Matrix of Numerical Features

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Drop non-numeric columns
numeric_data = dataset.select_dtypes(include=['int', 'float'])

# Impute missing values with the mean
numeric_data.fillna(numeric_data.mean(), inplace=True)

# Calculate correlation matrix
correlation_matrix = numeric_data.corr()

# Visualize correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```

# Team analysis

```python
# Team performances
# Calculate total runs scored by each team
team_runs = dataset.groupby('current_innings')['runs'].sum() #+ dataset.groupby('away_team')['runs'].sum()

# Display team performances
print("\nTeam performances:")
print("Total runs scored by each team:")
print(team_runs.sort_values(ascending=False))
```

# Comparisons between home and away teams

```python
# Comparisons between home and away teams
home_team_stats = dataset.groupby('home_team').agg({'batsman1_runs': 'sum', 'bowler1_wkts': 'sum'}).reset_index()
away_team_stats = dataset.groupby('away_team').agg({'batsman1_runs': 'sum', 'bowler1_wkts': 'sum'}).reset_index()

# Plot comparisons
plt.figure(figsize=(12, 6))
plt.bar(home_team_stats['home_team'], home_team_stats['batsman1_runs'], color='skyblue', label='Runs scored')
plt.bar(away_team_stats['away_team'], away_team_stats['batsman1_runs'], color='lightgreen', label='Runs scored')
plt.bar(home_team_stats['home_team'], -home_team_stats['bowler1_wkts'], color='orange', label='Wickets taken')
plt.bar(away_team_stats['away_team'], -away_team_stats['bowler1_wkts'], color='salmon', label='Wickets taken')
plt.xlabel('Team')
plt.ylabel('Runs / Wickets')
plt.title('Comparisons between Home and Away Teams')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```

# match_outcome

```python
import pandas as pd

# Assuming 'data' contains your dataset
# Create a DataFrame with selected columns
selected_columns = [
    'match_id', 'runs'
]

filtered_data = dataset[selected_columns].copy()

# Calculate total runs scored in each match
filtered_data['total_runs'] = filtered_data.groupby('match_id')['runs'].transform('sum')

# Define thresholds for classifying match outcomes
low_scoring_threshold = 200  # Define your threshold based on the dataset
high_scoring_threshold = 300  # Define your threshold based on the dataset

# Classify matches based on total runs scored
filtered_data['match_outcome'] = 'Moderate Scoring Match'
filtered_data.loc[filtered_data['total_runs'] < low_scoring_threshold, 'match_outcome'] = 'Low Scoring Match'
filtered_data.loc[filtered_data['total_runs'] > high_scoring_threshold, 'match_outcome'] = 'High Scoring Match'

# Drop duplicate rows based on match_id
filtered_data.drop_duplicates(subset='match_id', keep='first', inplace=True)

# Check the distribution of match outcomes
print(filtered_data['match_outcome'].value_counts())
```

```python
# Visualize the distribution of match outcomes
plt.figure(figsize=(8, 6))
filtered_data['match_outcome'].value_counts().plot(kind='bar', color=['skyblue', 'lightgreen', 'salmon'])
plt.title('Distribution of Match Outcomes')
plt.xlabel('Match Outcome')
plt.ylabel('Number of Matches')
plt.xticks(rotation=45)
plt.show()
```
